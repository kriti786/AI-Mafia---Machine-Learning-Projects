```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
dfx = pd.read_csv('D:\hhhhh\codeM\AI_Mafia\Week 2 - Assignments (Linear and Logistic Regression)\Linear Regression\Assignment 1\Training Data\Linear_X_Train.csv')
dfy = pd.read_csv('D:\hhhhh\codeM\AI_Mafia\Week 2 - Assignments (Linear and Logistic Regression)\Linear Regression\Assignment 1\Training Data\Linear_Y_Train.csv')

X = dfx.values
Y = dfy.values

X = (X-X.mean())/X.std()
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
model = LinearRegression()
```

```python
model.fit(X,Y)
```

```python
output = model.predict(X)
print(output)
```

```python
plt.scatter(X,Y,color='pink',label='data')
plt.plot(X,output,color='blue',label='prediction')
plt.legend()
plt.show()
```

```python
X_test=pd.read_csv("D:\hhhhh\codeM\AI_Mafia\Week 2 - Assignments (Linear and Logistic Regression)\Linear Regression\Assignment 1\Testing Data\Linear_X_Test.csv").values
```

```python
Y_test=model.predict(X_test)
```

```python
plt.plot(X_test,Y_test,color="Red")
plt.scatter(X,output,color="pink")
plt.show()
```

```python
model.score(X,Y)
```
